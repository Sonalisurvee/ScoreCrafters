```python
import pandas as pd
import pickle
import numpy as np
```

```python
matches = pickle.load(open('dataset_level1.pkl','rb'))
```

```python
match_id = []
batting_team = []
teams = []
balls_of_match = []
batsman = []
bowler = []
runs = []
city = []
venue = []
player_of_dismissed = []
player_of_match = []

count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    count+=1
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball:
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            balls_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            player_of_match.append(row['info.player_of_match'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
data = {
    'match_id': match_id,
    'teams': teams,
    'batting_team': batting_team,
    'balls': balls_of_match,
    'batsman': batsman,
    'bowler': bowler,
    'runs': runs,
    'player_of_dismissed': player_of_dismissed,
    'city': city,
    'venue': venue,
    'player_of_match' : player_of_match
}

delivery_df = pd.DataFrame(data)
```

```python
df = delivery_df.copy()
```

```python
df
```

```python
df['player'] = df['player_of_match'].apply(lambda x: x[0] if isinstance(x, list) else x)
df['player'] = df['player'].str.strip('[]')
df
```

```python
df.player.isnull().value_counts()
```

```python
df.drop(columns=['player_of_match'],inplace=True)
```

```python
df = df.rename(columns={'player': 'player_of_match'})
df
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
df['bowling_team'] = df.apply(bowl,axis=1)
df
```

```python
df.drop(columns=['teams'],inplace=True)
```

```python
team=[
    'India',
    'Pakistan',
    'South Africa',
    'Australia',
    'New Zealand',
    'England',
    'Sri Lanka',
    'Bangladesh',
    'Afghanistan',
    'West Indies'
]
```

```python
df = df[df['batting_team'].isin(team)]
df = df[df['bowling_team'].isin(team)]
```
